```python
from bs4 import BeautifulSoup as lula
import requests as eymael
import re
```

```python
htmlc = eymael.get('https://www.tempo.com/recife.htm')
htmlf = lula (htmlc.content)
```

```python
clima = htmlf.find_all('span', class_='temperatura')
data = htmlf.find_all('span', class_='cuando')
temp = []
dia = []
for i in clima:
  x = i.get_text()
  temp.append(x)
for j in data:
  x = j.get_text()
  dia.append(x)
for y in range(5):
  temp.pop(0)
for k in range(len(dia)):
  print(f"{dia[k]} - {temp[k]}")
```
